# Limpieza de datos

```python
#Eliminación de la columna Unnamed
df1=df1.drop(columns=["Unnamed: 0"])
df1
```

```python
#Renombramiento de columnas
NewCab = ['Anio', 'Tipo','Genero','Rating','Puntuacion', 'Pais_Origen','Idioma','Elenco','Director','Creador','Escritor','Duracion(min)']
NewCab
cab = np.array(NewCab)
cab
```

```python
#Contabilización de celdas vacías según cabecera
df1.isna().sum().sort_values()
```

```python
sns.heatmap(df1.isnull())
```

```python
df1.dropna(subset=["Genero","Tipo"], inplace=True)
df1
```

```python
# vamos a modificar el valor nullo por el valor medio
# calcular la duracion media x cada una de los tipos (episode, movie, tv mini series, tv movie, tv series, video movie)
"""la linea que divide la caja representa el valor medio
tv mini series: 185, movie: 97, video movie: 85, tv movie: 97, episode: 73, tv series: 40"""
sns.boxplot(x='Tipo', y="Duracion(min)", data=df1)
```

```python
#rellenar valore nulos con la media de los valores en Duracion
#tv mini series: 185, movie: 97, video movie: 85, tv movie: 97, episode: 73, tv series: 40"""
def llenar_duracion_media(columnas):
  duracion = columnas[0]
  tipo = columnas[1]
  if pd.isnull(duracion):
    if tipo == 'tv mini series':
      return 185
    elif tipo == 'movie':
      return 97
    elif tipo == 'video movie':
      return 85
    elif tipo == 'tv movie':
      return 97
    elif tipo == 'episode':
      return 73
    else:
      return 40
  else:
    return duracion

df1["Duracion(min)"] = df1[["Duracion(min)", "Tipo"]].apply(llenar_duracion_media, axis=1)
sns.heatmap(df1.isnull())
```

```python
# vamos a modificar el valor nullo por el valor medio
# calcular la rating media x cada una de los tipos (episode, movie, tv mini series, tv movie, tv series, video movie)
"""la linea que divide la caja representa el valor medio
tv mini series: 8.5, movie: 7.7, video movie: 7, tv movie: 7.8, episode: 7.4, tv series: 7.1"""
sns.boxplot(x='Tipo', y="Rating", data=df1)
```

```python
#rellenar valore nulos con la media de los valores en Rating
#tv mini series: 8.5, movie: 7.7, video movie: 7, tv movie: 7.8, episode: 7.4, tv series: 7.1"""
def llenar_rating_media(columnas):
  rating = columnas[0]
  tipo = columnas[1]
  if pd.isnull(rating):
    if tipo == 'tv mini series':
      return 8.5
    elif tipo == 'movie':
      return 7.7
    elif tipo == 'video movie':
      return 7
    elif tipo == 'tv movie':
      return 7.8
    elif tipo == 'episode':
      return 7.4
    else:
      return 7.1
  else:
    return rating

df1["Rating"] = df1[["Rating", "Tipo"]].apply(llenar_rating_media, axis=1)
sns.heatmap(df1.isnull())
```

```python
media = df1["Puntuacion"].mean()
mediana = df1["Puntuacion"].median()
moda = df1["Puntuacion"].mode()

print(media, mediana, moda)
```

```python
#Rellenar columna puntuacion random dentro de un rango 18 - 75773
import random
def llenar_puntuacion_media(columna):
    puntuacion = columna[0]
    if pd.isnull(puntuacion):
        return random.randint(18, 75773)
    else:
        return puntuacion
    
df1["Puntuacion"] = df1[["Puntuacion", "Tipo"]].apply(llenar_puntuacion_media, axis=1)
sns.heatmap(df1.isnull())
```
